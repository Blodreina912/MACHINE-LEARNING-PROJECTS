## Generate synthetic data

### Subtask:
Create a dataset with a non-linear relationship between X and y using a quadratic function with added noise.


**Reasoning**:
Generate synthetic data with a non-linear relationship using a quadratic function with added noise and reshape X.



```python
import numpy as np

np.random.seed(0)
X = 10 * np.random.rand(100) - 5
y = X**2 + np.random.randn(100) * 5
X = X.reshape(-1, 1)
```

## Transform features

### Subtask:
Apply polynomial feature transformation to increase the dimensionality of the feature space.


**Reasoning**:
Apply polynomial feature transformation to increase the dimensionality of the feature space.



```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
```

## Train a linear regression model

### Subtask:
Fit a linear regression model to the transformed features.


**Reasoning**:
Import the LinearRegression class, instantiate it, and fit the model to the transformed features and target variable.



```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_poly, y)
```

## Visualize results

### Subtask:
Plot the original data points and the fitted polynomial regression curve.


**Reasoning**:
Plot the original data points and the fitted polynomial regression curve.



```python
import matplotlib.pyplot as plt

plt.scatter(X, y, label='Original data')
X_plot = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
X_plot_poly = poly.transform(X_plot)
y_plot = model.predict(X_plot_poly)
plt.plot(X_plot, y_plot, color='red', label='Polynomial regression curve')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Polynomial Regression Fit')
plt.legend()
plt.show()
```

## Evaluate performance

### Subtask:
Calculate and display the Mean Squared Error (MSE) to assess the model's performance.


**Reasoning**:
Calculate the Mean Squared Error using the actual and predicted values and print the result.



```python
from sklearn.metrics import mean_squared_error

y_pred = model.predict(X_poly)
mse = mean_squared_error(y, y_pred)
print(f'Mean Squared Error (MSE): {mse}')
```

## Summary:

### Data Analysis Key Findings

*   Polynomial features of degree 2 were successfully created from the original feature X, expanding the feature space.
*   A Linear Regression model was successfully trained on the polynomial transformed features.
*   A visualization was generated showing the original data points and the fitted polynomial regression curve, demonstrating that the model effectively captures the non-linear relationship.
